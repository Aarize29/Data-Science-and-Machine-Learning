```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

#for current_ax in axes:
#    current_ax.plot(x,y)

axes[0].plot(x,y)
axes[1].plot(y,x)


plt.tight_layout()
```

```python
# Figure size and dpi
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#to save
fig.savefig('my_picture.png')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='square')
ax.plot(x,x**3,label='cube')


ax.legend(loc='best') #loc-location
```
