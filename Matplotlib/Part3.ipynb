```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
#plot appearance

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3, linestyle='dashed', marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=4,markeredgecolor='red')  #color me kuch bhi dal do

ax.set_xlim([0,4])
ax.set_ylim([0,3])
```
